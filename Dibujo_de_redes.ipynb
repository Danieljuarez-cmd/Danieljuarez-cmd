La libreria sirve para crear, analizar y manipular redes comolejas de mnera facil.

```python
G=nx.DiGraph()#Grafo dirigido
G.add_edge("A","B")
G.add_edge("B","C")
G.add_edge("B","D")
G.add_edge("A","C")
G.add_edge("C","D")
G
nx.draw_networkx(G)

```
