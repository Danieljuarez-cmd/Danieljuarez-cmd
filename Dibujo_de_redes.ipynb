```python
import networkx as nx
```

La libreria sirve para crear, analizar y manipular redes comolejas de mnera facil.

En la siguiente parte ponemos primero una red vacia para crear directamente la grafica

```python
G=nx.DiGraph()#Grafo dirigido
```

A continuacion se crean los arcos que unen a un nodo con otro

Representa la red nodo por nodo

```python

G.add_edge("A","B")
G.add_edge("B","C")
G.add_edge("B","D")
G.add_edge("A","C")
G.add_edge("C","D")

nx.draw_networkx(G, node_color="green")#Muestra la red con ptro color

```
